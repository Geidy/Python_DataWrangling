```python
import pandas as pd
import matplotlib.pylab as plt
```

```python
file_path="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
print(headers)
```

```python
#there were two options to call the header
#df = pd.read_csv(file_path, names = headers) --> because I have created a header list
#df = pd.read_csv(file_path, header = 1) or header=0 if didn't want to assign any header to the dataset
df = pd.read_csv(file_path, names = headers)
```

```python
#checking how the set and header assignment looks like 
df.head(10)
```

```python
Identify and handling missing values 
Convert "?" to NaN
.replace(A, B, inplace = True)
to replace A by B
#importing numpy
```

```python
import numpy as np
df.replace( "?", np.nan, inplace = True)
df.head(20)
```

```python
Evaluating Missing Data

.isnull()
.notnull()

The output a boolean value indicating whether the value that passed into the argument  fact missing data.
```

```python
missing_data = df.isnull()
missing_data.head(20)
#True means the value is a missing value
#while False means value is not missing
```

```python
#Counting Missing values on each column
#Dealing with Missing Values
We could eather drop or replace the data missing
```

```python
Replace by mean:
Replace by frequency:
Drop the whole row:
# "price": 4 missing data, simply delete the whole row
Reason: You want to predict price. You cannot use any data entry without price data for prediction; 
    therefore any row now without price data is not useful 
```

```python
#Calculating mean values for calculated losses columns 

avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized-losses:", avg_norm_loss)
```

```python
#Replace NAN with mean value in normalized losses columns

df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
```

```python
# Calculating Mean value for bore column
avg_bore=df['bore'].astype('float').mean(axis=0)
print("Average of bore:", avg_bore)
```

```python
#Replace "NaN" with the mean value in the "bore" column

df["bore"].replace(np.nan, avg_bore, inplace=True)
```

```python
#Correct data format
.dtype() to check the data type
.astype() to change the data type
```

```python
df.dtypes
```

```python
#Convert data types to proper format
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```

```python
#Checking after conversion
df.dtypes
```

```python
Data Standardization
Standardization is the process of transforming data into a common format, 
allowing you to make a
meaningful comparison.

Transform mpg to L/100km:
    
    Use this formula for unit conversion:
L/100km = 235 / mpg
```

```python
# Convert mpg to L/100km by mathematical operation (235 divided by mpg)
df['city-L/100km'] = 235/df["city-mpg"]

# check your transformed data 
df.head()
```

```python
Data Normalization
Why normalization?
Normalization is the process of transforming values of several variables into a similar range. Typical normalizations include
scaling the variable so the variable average is 0
scaling the variable so the variance is 1
scaling the variable so the variable values range from 0 to 1
```

```python
# replace (original value) by (original value)/(maximum value)
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
```

```python
#Normalizing High

df['height'] = df['height']/df['height'].max() 

# show the scaled columns
df[["length","width","height"]].head()
```

```python
Binning

Binning is a process of transforming continuous numerical variables into discrete categorical 'bins' 
for grouped analysis.
```

```python
#Converting to correct format

#df["horsepower"]=df["horsepower"].astype(int, copy=True) # getting error because of nan values on Horsepower

#Dropping NAN values and converting the format
df["horsepower"] = pd.to_numeric(df["horsepower"], errors="coerce")  # Ensure numeric type
df = df.dropna(subset=["horsepower"])  # Drop NaNs
df["horsepower"] = df["horsepower"].astype(int)  # Convert to int
```

```python
Plotting Histogram of horsepower to see the distribution of horsepower.
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df["horsepower"])

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
Finding 3 bins of equal size bandwidth by using Numpy's linspace(start_value, end_value, numbers_generated function.
```

```python
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins
```

```python
Setting Groups name
```

```python
group_names = ['Low', 'Medium', 'High']
```

```python
Apply the function "cut" to determine what each value of df['horsepower'] belongs to.
```

```python
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True )
df[['horsepower','horsepower-binned']].head(20)
```

```python
Number of vehicles on each bin
```

```python
df["horsepower-binned"].value_counts()
```

```python
Plotting Distribution of each bin
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, df["horsepower-binned"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```
